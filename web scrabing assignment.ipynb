```python
import requests
```

```python
url=('https://footballdatabase.com/ranking/asia/1')
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select(".limittext")
```

```python
name
```

```python
countries=soup.select(".sm_logo-name")
```

```python
countries
```

```python
points=soup.select(".text-left+ .rank")
```

```python
points
```

```python
import pandas
```

```python
football=pandas.DataFrame(name,columns=['club name'])
```

```python
football
```

```python

football["country"]=countries
```

```python
football["points"]=points
```
